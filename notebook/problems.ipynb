### Exercise 1

We will use NumPy to obtain information to describe statistically.

- Generate an array of 100 elements following a normal distribution.
- Generate an array of 100 elements following a chi-square distribution with 3 degrees of freedom.
- Calculate the main metrics and statistical measures that best describe the two vectors.

````python
# TODO
import numpy as np
import statistics as stats
from scipy.stats import skew
from scipy.stats import kurtosis


#Generate an array of 100 elements following a normal distribution
normal = np.random.normal(size = 100)
print(normal)

#Generate an array of 100 elements following a chi-square distribution with 3 degrees of freedom
chi = np.random.chisquare(df = 3, size = 100)
print(chi)

#Calculate the main metrics and statistical measures that best describe the two vectors.

#Medidas de tendencia central

    # Media
print(f'Normal mean:{stats.mean(normal)}')
print(f'chi mean:{stats.mean(chi)}')

    # Mediana
print(f'Normal median: {stats.median(normal)}')
print(f'chi median: {stats.median(chi)}')

    # Moda
print(f'Normal Mode: (stats.mode(normal))')
print(f'Chi mode: {stats.mode(chi)}')

#Medidas de dispersion

    #Rango
rango_normal = max(normal) - min(normal)
rango_chi = max(chi) - min(chi)
print(f"Normal range: {rango_normal}")
print(f"Normal chi: {rango_chi}")

    #Desviacion standar
standar_normal = stats.stdev(normal)
standar_chi = stats.stdev(chi)

    #varianza
varianza_normal = stats.variance(normal)
varianza_chi = stats.variance(chi)

# Shape measures

    # Skewness
skew_normal = skew(normal)
skew_chi = skew(chi)

print(f"Normal skewness: {skew_normal}")
print(f"Chi skewness: {skew_chi}")

    # Kurtosis
kurt_normal = kurtosis(normal)
kurt_chi = kurtosis(chi)
print(f"Normal kurtosis: {kurt_normal}")
print(f"Chi kurtosis: {kurt_chi}")
````

### Exercise 2

Write a Python program to calculate the standard deviation of the following data:

```py
data = [4, 2, 5, 8, 6]
```

````python
# TODO
import math
import sys

# Define the standard deviation function

def sd_calc(data):
    n = len(data)

    if (n <= 1):
        return 0.0

    mean, sd = avg_calc(data), 0.0

    for d in data:
        sd += (float(d) - mean) ** 2
    sd = math.sqrt(sd / float(n - 1))

    return sd

def avg_calc(data):
    n, mean = len(data), 0.0

    if (n <= 1):
        return data[0]

    for d in data:
        mean = mean + float(d)

    mean = mean / float(n)
    return mean


data = [4, 2, 5, 8, 6]
print(f"Sample Data: {data}")
print(f"Standard Deviation: {sd_calc(data)}")
````
